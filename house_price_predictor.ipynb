```python
import pandas as pd
df = pd.read_csv('data/housing.csv')
df
```

```python
df.describe()
```

```python
import seaborn as sns

# plot
sns.scatterplot(data=df, x="median_income", y="median_house_value")
```

```python
sns.scatterplot(data=df, x="population", y="median_house_value")
```

```python
from sklearn import linear_model 

model = linear_model.LinearRegression()
```

```python
model.fit(X=df[["median_income"]], y =df[["median_house_value"]])
```

```python
model.intercept_
```

```python
12*model.coef_+ model.intercept_
```

```python
model.coef_
```

```python
model.predict(X=[[12],[1],[100]])
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x="median_income", y="median_house_value", color = "orange")
plt.plot(df[["median_income"]],model.predict(df[["median_income"]]), color = "blue",linewidth = 3)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error


mean_absolute_error(df[["median_house_value"]],model.predict(df[["median_income"]]))
```

```python
mean_squared_error(df[["median_house_value"]],model.predict(df[["median_income"]]))
```
